```python
def check_isnan (data, pourcent):
    columns_null = []
    for i in data.columns:
        if data[i].isnull().sum() > data.shape[0]*pourcent/100:
            columns_null.append(i)
    return columns_null

        
```

```python
# Premier clean : virer les colonnes qui sont renseigné pour aucun des produit
# En passant, il est étrange de voir des colonnes comme '-glucose_100g' totalement non renseignés
check_isnan(data, 90)
```

```python
def delete_isnan (data,pourcent):
    to_delete = check_isnan(data,pourcent)
    data.drop([i for i in to_delete], axis = 1, inplace = True)
    return data
```

```python
# Vérifions si les colonnes de float sont bien remplies de float (on sait jamais)
def check_type_column (data):
    liste_a_check = []
    for i in data.columns:
        if (data[i].dtype==float):
            if not np.array_equal(data[i], data[i].astype(data[i].dtype), equal_nan=True):
                liste_a_check.append(i)
    return liste_a_check
```

```python
# Je crée une fonction permettant d'afficher des données à partir d'une liste érite sous la forme
# [
# ['nom colone1',[liste1 d'indices]] ,
# ['nom colone2',[liste2 d'indices] ,
# ['nom colone3',[liste3 d'indices] ,
# ...
# ]
# Prenant en entrée le dataset, la liste, et l'indice dans cet liste à afficher
def comparer_valeurs_intrigantes_ou_aberrantes (data,liste,i):
    if len(liste)>= i:
        columns = ['product_name']
        columns.append(liste[i][0])
        df = data.loc[:,columns].iloc[liste[i][1],:]
        df_new_line = pd.DataFrame([['moyenne de lensemble de la colonne',data[columns[1]].mean()]], columns=[columns[0], columns[1]] )
        df = pd.concat([df,df_new_line])
        return df
    else :
        return 'indice proposé trop grand, il ny a plus de valeurs intrigantes'       
```

```python
# J'appelle une valeur intrigante une valeur qui est anormalement éloigné de la moyenne.
# La fonction permet donc de lister selon la colonne les indices ayant des valeurs hors [m - x*eq type , m + x.eq] pour x.
# format de sortie = format d'entrée de la fonction ci dessus
def valeurs_intrigantes(data,nb_eq_type):
    liste_a_check = []
    for i in data.columns:
        if data[i].dtype == float:
            index = data.loc[(data[i] > data[i].mean() + nb_eq_type*data[i].std()) | (data[i] < data[i].mean() - nb_eq_type*data[i].std()), i].index.tolist()
            if not len(index) == 0:
                liste_a_check.append([i,index])
    return liste_a_check
        
```

```python
VI = valeurs_intrigantes(data,100)
VI
```

```python
comparer_valeurs_intrigantes_ou_aberrantes (data,VI,4)
```

```python
# On va appeller une valeur aberante, une valeur dans une catégorie "**** par 100g" qui contiens une 
# valeur plus grande que 100. Logiquement impossible de contenir plus de 100g par 100g.
# Exception importante : si les valeurs ne sont pas exprimés en gramme (Kj, Kcal, mg...) alors la valeur n'est plus abérrante
# Donc on ne considèrera que les valeurs qui ont une moyenne plus petite que 100, mais qui ont malgré tout des valeurs > 100
# Sortira une liste de format utilisé dans la fonction d'affichage utilisé ci dessus
def valeurs_aberrante(data): 
    liste_a_check = []
    for i in data.columns:
        if (data[i].dtype == float):
            if ('100g' in i):
                idx = data.loc[(data[i] > 100) | (data[i]<0), i].index.tolist()
                if not len(idx) == 0:
                    liste_a_check.append([i,idx])
    return liste_a_check
```

```python
# Créons vite fais une fonction récupérant juste les noms des colonnes concernés dans la liste de même type que précédement.
def recup_colonne(data, liste):
    nom_columns = []
    for i in liste :
        nom_columns.append(i[0])
    return nom_columns
```

```python
# Très vite on comprends que cetraines valeurs ne sont pas si intrigante comme la quantité produit, forcément
# Bien plus grande pour des produit comme le nutella que pour un obscur fromage corse.
recup_colonne(data, VI)
```

```python
# Utilisons la liste crée pour remplacer ces valeurs soit par la moyenne, soit par NaN, quand on en aura besoin
# a voir si ça existe pas
def remplacer_data_aberrante_par (data,liste,changement):
    for i in liste:
        if changement == 'mean':
            new_value = data[i[0]].mean()
        elif changement == 'nan':
            new_value = np.nan
        else :
            return 'Wrong value for changement'
        for k in i[1]:
            data.loc[k,i[0]] = new_value
    return data

```

```python
# Nous allons developper l'idée d'une application pour les personnes diabétique. Celles ci doivent bien sur se nourrir,
# Mais leur alimentation doit éviter des produit ayant certaines des quantités de notre dataset trop importantes.
# Nous allons donc proposer une application qui, au scan d'un article, propose des informations chiffrés sur la composition,
# une comparaison avec une valeur définie comme étant correcte pour les diabétiques, et l'information principale :
# "Bon pour une personne diabétique en quantité raisonable" ou
# "Bon pour une personne diabétique en petite quantité" ou
# "Mauvais pour une personne diabétique même en petite quantité"
# Accompagnant ces deux dernière information, il y aura l'indication du/des composants empêchant de considerer l'article comme
# Etant de la première catégorie pour que l'utilisateur puisse compenser en achetant d'autres produit ayant moins de cette 
# quantité.
# Accompagnant la première ligne des 3 proposés, on aura au contraire l'information de quelle quantité est le moins présente
# Pour laquel l'utilisateur pourra se permettre d'acheter un article en ayant un peu plus.
# 
# Une idée finale d'application à l'aide de ces information serait de scanner l'ensemble des aliments d'un repas pour vérifier 
# l'équilibre de celui ci pour une personne diabétique.
#
# Nous allons donc avoir besoin des informations suivantes : les glucides, les sucres, les fibres, le gras (acide gras saturé 
# ou pas) , le cholesterol, l'apport energétique, le sodium et le sel.
```

```python
# D'après le site "Journal des femmes", ayant classé les produit ayant le plus de tel ou tel élément dans sa composition, 
# J'établis un dictionnaire de valeur limite :
d = {'sugars_100g' : 100.01, 
     'fiber_100g' : 45, 
     'energy-kcal_100g' : 1000, 
     'fat_100g' : 100.01, 
     'saturated-fat_100g' : 100.01, # Si je met 100, il me retire également les valeurs égal à exactement 100, j'ajoute donc .01 pour éviter le problème.
     'cholesterol_100g' : 3.5 ,
     'salt_100g' : 100.01 ,
     'sodium_100g' : 50}
```

```python
# On retire toutes les valeurs ne respectant pas ces limites maximales (que je rappelle, sont des valeurs recherchés comme
# etant les valeurs maximales qu'un aliment peut posséder selon https://sante.journaldesfemmes.fr/calories/classement/aliments/cholesterol)
for i in Liste_quantitative:
    data_final.loc[data_final[i] >= d[i], i] = np.nan
```

```python
data_final.describe()
# Etape suivante : remplir les trous. On va procéder à un KNN avec k = 3
```

```python
# Nous avons maintenant un set de données crédible, mais très troués. On va donc remplir les trous en appliquant un Knn.
# Pour celà, nous devons avant tout centrer et normaliser nos valeurs, pour que le calcul de la distance ai un sens lors
# du Knn.
x = data_final.loc[: , data_final.columns.isin(Liste_quantitative)]
scaler = StandardScaler() #Chargement du centrage réduction
scaler.fit(x)
X_cr = scaler.transform(x)
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
X_knn = imputer.fit_transform(X_cr)
```

```python
# On décrentre et "déréduit" nos valeurs pour les entrer dans le tableau :
data_final.loc[: , data_final.columns.isin(Liste_quantitative)] = scaler.inverse_transform(X_knn)
```
